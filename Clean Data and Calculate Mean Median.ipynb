```python
import pandas as pd
import numpy as np
import os
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

#Reading Excel File
housing= pd.read_excel('C:/Users/Arohi Ganatra/Desktop/BostonHousing.xlsx', sheet_name='Data')


#Function to highlight all non-numeric & blank fields
def highlight_values(value):
    if pd.isna(value) or isinstance(value, str) or value == "nan":
        return 'background-color: yellow'
    return ''

#Function for identifying outliers
def identify_outlier(row):
    
    value = row['PTRATIO']
    
    if pd.isnull(value) or value == "" or not isinstance(value, (int,float, np.number)):
        return 'a'  # Non-numeric value
    
    elif value < 3 or value >= 60:
        return 'b'  # Shift in decimal place or data entry error
    
    elif value > 40 and value < 60:
        return 'c'  # Genuine case of outlier
    else:
        return ""

#Adding new column by call outlier function    
housing['Outlier Cause']= housing.apply(identify_outlier, axis=1)

#Making column list to exclude newly added column as well as excluding PTRATIO column as per question
t = list(housing.columns)
t.remove('PTRATIO')
t.remove('Outlier Cause')

#Calling highlight_values function & using applymap method of Styler library highlighting required fields
housing_highlighted=housing.style.applymap(highlight_values, subset=pd.IndexSlice[:,t])

#Display
housing_highlighted

```

```python
#Removing all rows with non numeric values & blank fields........OMISSION

print('')
print('OMISSION')
print('')
housing_omission= pd.read_excel('C:/Users/Arohi Ganatra/Desktop/BostonHousing.xlsx', sheet_name='Data')

a=housing_omission.shape
print('Before Omission')
print('Rows,Columns')
print(a)

#Replacing non-numeric values to NaN
housing_omission = housing_omission[housing_omission.applymap(lambda x: isinstance(x, (int, float)))]
#Dropping all rows with NaN values
housing_omitted= housing_omission.dropna(axis=0, inplace =True, how ='any')


housing_omission.to_excel('C:/Users/Arohi Ganatra/Desktop/BostonHousing_omittedData.xlsx', index= False)
a=housing_omission.shape
print('')
print('After Omission')
print('Rows,Columns')
print(a)
```

```python
#Imputation procedure replacing all the non numeric fields with mean of the column data.......IMPUTATION

print('')
print('IMPUTATION')
print('')
housing_impute= pd.read_excel('C:/Users/Arohi Ganatra/Desktop/BostonHousing.xlsx', sheet_name='Data')

# Replacing all non numeric values to NaN
housing_impute = housing_impute[housing_impute.applymap(lambda x: isinstance(x, (int, float)))]

#Filling NaN values as column mean
housing_impute=housing_impute.fillna(housing_impute.mean())
housing_impute.to_excel('C:/Users/Arohi Ganatra/Desktop/BostonHousing_imputedData.xlsx', index= False)
housing_impute

```
